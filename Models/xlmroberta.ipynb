```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import pickle
import os
import time
import torch
import random
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.metrics import precision_recall_fscore_support
from sklearn.preprocessing import LabelEncoder
from transformers import XLMRobertaTokenizer, AdamW
from transformers import get_linear_schedule_with_warmup
from transformers import AutoTokenizer, AutoModel
import torch.nn as nn
import torch.nn.functional as F
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')

# Set random seed for reproducibility
RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)
torch.manual_seed(RANDOM_STATE)
random.seed(RANDOM_STATE)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(RANDOM_STATE)

# Check for GPU availability
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Preprocess Sinhala text
def preprocess_sinhala_text(text):
    """
    Preprocess Sinhala text by removing non-Sinhala characters and extra whitespace
    """
    if isinstance(text, str):
        # Keep only Sinhala Unicode range (0D80-0DFF) and basic punctuation
        text = re.sub(r'[^\u0D80-\u0DFF\s.,!?]', '', text)
        # Remove extra whitespace
        text = re.sub(r'\s+', ' ', text).strip()
        return text
    return ""

# Simple text augmentation for Sinhala
def augment_text(text, prob=0.3):
    """Simple text augmentation by reordering words slightly"""
    if not isinstance(text, str) or random.random() > prob:
        return text
        
    words = text.split()
    if len(words) <= 3:
        return text
        
    # Randomly swap some adjacent words
    for i in range(len(words) - 1):
        if random.random() < 0.3:
            words[i], words[i+1] = words[i+1], words[i]
            
    return " ".join(words)

# Implementation of Focal Loss
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=2, reduction='mean'):
        super(FocalLoss, self).__init__()
        self.gamma = gamma
        self.reduction = reduction
        self.alpha = alpha  # Weight for each class
        
    def forward(self, inputs, targets):
        ce_loss = F.cross_entropy(inputs, targets, reduction='none', weight=self.alpha)
        pt = torch.exp(-ce_loss)
        focal_loss = (1 - pt) ** self.gamma * ce_loss
        
        if self.reduction == 'mean':
            return focal_loss.mean()
        elif self.reduction == 'sum':
            return focal_loss.sum()
        else:
            return focal_loss

# Create a PyTorch Dataset for XLM-RoBERTa
class MentalHealthDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_length=128, augment=False):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_length = max_length
        self.augment = augment
        
    def __len__(self):
        return len(self.texts)
    
    def __getitem__(self, idx):
        text = str(self.texts[idx])
        
        # Apply augmentation in training mode
        if self.augment:
            text = augment_text(text)
            
        label = self.labels[idx]
        
        # Tokenize the text
        encoding = self.tokenizer(
            text,
            add_special_tokens=True,
            max_length=self.max_length,
            padding='max_length',
            truncation=True,
            return_tensors='pt'
        )
        
        # Convert to required format
        item = {
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'label': torch.tensor(label, dtype=torch.long)
        }
        
        return item

# Define an improved model based on XLM-RoBERTa
class ImprovedXLMRobertaAdapter(nn.Module):
    def __init__(self, num_classes, freeze_base=True, unfreeze_layers=1, adapter_size=64, dropout=0.3):
        super(ImprovedXLMRobertaAdapter, self).__init__()
        
        # Load pre-trained XLM-RoBERTa
        self.base_model = AutoModel.from_pretrained("xlm-roberta-base")
        
        # Get hidden size from model config
        hidden_size = self.base_model.config.hidden_size
        
        # Freeze base model params to prevent overfitting
        if freeze_base:
            for param in self.base_model.parameters():
                param.requires_grad = False
            
            # Unfreeze the last N transformer layers
            for i in range(1, unfreeze_layers + 1):
                if i <= len(self.base_model.encoder.layer):
                    for param in self.base_model.encoder.layer[-i].parameters():
                        param.requires_grad = True
                    print(f"Unfrozen layer: {len(self.base_model.encoder.layer) - i + 1}")
        
        # Add adapter layers with bottleneck architecture
        self.adapter_down = nn.Linear(hidden_size, adapter_size)
        self.adapter_up = nn.Linear(adapter_size, hidden_size)
        self.activation = nn.GELU()
        self.layer_norm = nn.LayerNorm(hidden_size)
        self.dropout = nn.Dropout(dropout)
        
        # Classification layers
        self.classifier_hidden = nn.Linear(hidden_size, hidden_size // 2)
        self.classifier = nn.Linear(hidden_size // 2, num_classes)
        
    def forward(self, input_ids, attention_mask):
        # Get the base model outputs
        outputs = self.base_model(input_ids=input_ids, attention_mask=attention_mask)
        sequence_output = outputs.last_hidden_state
        
        # Get CLS token embedding
        cls_output = sequence_output[:, 0, :]
        
        # Apply adapter (bottleneck architecture)
        adapter_output = self.adapter_down(cls_output)
        adapter_output = self.activation(adapter_output)
        adapter_output = self.adapter_up(adapter_output)
        
        # Residual connection and layer normalization
        output = self.layer_norm(cls_output + adapter_output)
        
        # Apply classification layers with dropout
        output = self.dropout(output)
        output = self.classifier_hidden(output)
        output = self.activation(output)
        output = self.dropout(output)
        logits = self.classifier(output)
        
        return logits

# Early stopping implementation
class EarlyStopping:
    def __init__(self, patience=5, min_delta=0.001, restore_best_weights=True):
        self.patience = patience
        self.min_delta = min_delta
        self.restore_best_weights = restore_best_weights
        self.best_score = None
        self.best_weights = None
        self.counter = 0
        self.stopped_epoch = 0
        
    def __call__(self, score, model):
        if self.best_score is None:
            self.best_score = score
            self.best_weights = {k: v.clone() for k, v in model.state_dict().items()}
            return False
        
        if score > self.best_score + self.min_delta:
            self.best_score = score
            self.counter = 0
            if self.restore_best_weights:
                self.best_weights = {k: v.clone() for k, v in model.state_dict().items()}
            return False
        
        self.counter += 1
        if self.counter >= self.patience:
            self.stopped_epoch = self.counter
            if self.restore_best_weights:
                model.load_state_dict(self.best_weights)
            return True
        
        return False

# Training function with gradient accumulation
def train_epoch(model, data_loader, optimizer, scheduler, device, 
                loss_fn, accumulation_steps=4):
    model.train()
    total_loss = 0
    correct_predictions = 0
    total_predictions = 0
    
    # Use tqdm for progress bar
    progress_bar = tqdm(data_loader, desc="Training")
    
    optimizer.zero_grad()  # Zero gradients initially
    
    for batch_idx, batch in enumerate(progress_bar):
        # Move batch to device
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['label'].to(device)
        
        # Forward pass
        logits = model(input_ids=input_ids, attention_mask=attention_mask)
        
        # Calculate loss
        loss = loss_fn(logits, labels)
        
        # Scale loss by accumulation steps
        loss = loss / accumulation_steps
        
        # Backward pass
        loss.backward()
        
        # Only update parameters after accumulating gradients
        if (batch_idx + 1) % accumulation_steps == 0 or (batch_idx + 1) == len(data_loader):
            # Clip gradients to prevent exploding gradients
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            
            # Update parameters
            optimizer.step()
            scheduler.step()
            optimizer.zero_grad()
        
        # Update metrics
        total_loss += loss.item() * accumulation_steps  # Scale loss back up for reporting
        
        _, preds = torch.max(logits, dim=1)
        correct_predictions += torch.sum(preds == labels).item()
        total_predictions += len(labels)
        
        # Update progress bar
        progress_bar.set_postfix({'loss': loss.item() * accumulation_steps})
    
    # Calculate metrics
    epoch_loss = total_loss / len(data_loader)
    epoch_acc = correct_predictions / total_predictions
    
    return epoch_loss, epoch_acc

# Validation function
def eval_model(model, data_loader, device, loss_fn):
    model.eval()
    total_loss = 0
    correct_predictions = 0
    total_predictions = 0
    all_preds = []
    all_labels = []
    
    with torch.no_grad():
        for batch in tqdm(data_loader, desc="Evaluating"):
            # Move batch to device
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['label'].to(device)
            
            # Forward pass
            logits = model(input_ids=input_ids, attention_mask=attention_mask)
            
            # Calculate loss
            loss = loss_fn(logits, labels)
            
            # Update metrics
            total_loss += loss.item()
            
            _, preds = torch.max(logits, dim=1)
            correct_predictions += torch.sum(preds == labels).item()
            total_predictions += len(labels)
            
            # Collect predictions and labels for metrics calculation
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
    
    # Calculate metrics
    epoch_loss = total_loss / len(data_loader)
    epoch_acc = correct_predictions / total_predictions
    
    return epoch_loss, epoch_acc, all_preds, all_labels

# Plot metrics during training
def plot_training_metrics(train_losses, val_losses, train_accs, val_accs, title='Training Metrics'):
    plt.figure(figsize=(16, 6))
    
    # Plot loss
    plt.subplot(1, 2, 1)
    plt.plot(train_losses, 'b-', label='Training Loss')
    plt.plot(val_losses, 'r-', label='Validation Loss')
    plt.title(f'{title} - Loss', fontsize=14)
    plt.xlabel('Epoch', fontsize=12)
    plt.ylabel('Loss', fontsize=12)
    plt.grid(True)
    plt.legend(loc="upper right", fontsize=12)
    
    # Calculate gap for loss
    if len(train_losses) > 0 and len(val_losses) > 0:
        gap_loss = train_losses[-1] - val_losses[-1]
        plt.annotate(f'Loss gap: {abs(gap_loss):.4f}',
                    xy=(0.5, 0.05), 
                    xycoords='axes fraction',
                    horizontalalignment='center',
                    verticalalignment='bottom',
                    fontsize=12, 
                    bbox=dict(boxstyle="round,pad=0.3", 
                            fc="white", ec="gray", alpha=0.8))
    
    # Plot accuracy
    plt.subplot(1, 2, 2)
    plt.plot(train_accs, 'b-', label='Training Accuracy')
    plt.plot(val_accs, 'r-', label='Validation Accuracy')
    plt.title(f'{title} - Accuracy', fontsize=14)
    plt.xlabel('Epoch', fontsize=12)
    plt.ylabel('Accuracy', fontsize=12)
    plt.grid(True)
    plt.legend(loc="lower right", fontsize=12)
    
    # Calculate gap for accuracy
    if len(train_accs) > 0 and len(val_accs) > 0:
        gap_acc = train_accs[-1] - val_accs[-1]
        plt.annotate(f'Accuracy gap: {gap_acc:.4f}',
                    xy=(0.5, 0.05), 
                    xycoords='axes fraction',
                    horizontalalignment='center',
                    verticalalignment='bottom', 
                    fontsize=12, 
                    bbox=dict(boxstyle="round,pad=0.3", 
                            fc="white", ec="gray", alpha=0.8))
    
    plt.tight_layout()
    filename = title.lower().replace(' ', '_') + '.png'
    plt.savefig(filename)
    plt.show()

# Get metrics from predictions
def get_metrics(y_true, y_pred):
    accuracy = accuracy_score(y_true, y_pred)
    precision, recall, f1, _ = precision_recall_fscore_support(
        y_true, y_pred, average='weighted')
    
    metrics = {
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1': f1
    }
    
    return metrics

# Prediction function
def predict_mental_health(text, tokenizer, model, label_encoder, device, max_length=128):
    """
    Predict mental health category from text
    """
    # Preprocess the text
    preprocessed_text = preprocess_sinhala_text(text)
    
    # Tokenize
    encoding = tokenizer(
        preprocessed_text,
        add_special_tokens=True,
        max_length=max_length,
        padding='max_length',
        truncation=True,
        return_tensors='pt'
    )
    
    # Move to device
    input_ids = encoding['input_ids'].to(device)
    attention_mask = encoding['attention_mask'].to(device)
    
    # Set model to evaluation mode
    model.eval()
    
    # Make prediction
    with torch.no_grad():
        logits = model(input_ids=input_ids, attention_mask=attention_mask)
        probabilities = torch.softmax(logits, dim=1).cpu().numpy()[0]
        prediction = torch.argmax(logits, dim=1).cpu().numpy()[0]
    
    # Get the predicted class
    predicted_class = label_encoder.inverse_transform([prediction])[0]
    
    # Return results
    result = {
        'class': predicted_class,
        'confidence': float(probabilities[prediction]),
        'probabilities': {
            label_encoder.classes_[i]: float(probabilities[i])
            for i in range(len(label_encoder.classes_))
        }
    }
    
    return result

# Main function to train the model with improved approach
def train_improved_transformer(data_path, output_dir='./sinhala_mental_health_model',
                              unfreeze_layers=2, adapter_size=128, dropout=0.4, 
                              epochs=15, batch_size=16, max_length=128, 
                              learning_rate=1e-5, use_focal_loss=True, 
                              use_augmentation=True):
    """
    Train an improved XLM-RoBERTa adapter model for Sinhala mental health classification
    """
    # Start timing
    start_time = time.time()
    
    # Create output directory
    os.makedirs(output_dir, exist_ok=True)
    
    # Load and preprocess data
    print("Loading dataset...")
    df = pd.read_csv(data_path)
    print(f"Dataset shape: {df.shape}")
    print(df['label'].value_counts())
    
    # Apply preprocessing
    df['preprocessed_text'] = df['cleaned_text'].apply(preprocess_sinhala_text)
    
    # Encode the labels
    label_encoder = LabelEncoder()
    df['encoded_label'] = label_encoder.fit_transform(df['label'])
    print(f"Classes: {label_encoder.classes_}")
    
    # Split the data with stratification
    X_train, X_val, y_train, y_val = train_test_split(
        df['preprocessed_text'], 
        df['encoded_label'],
        test_size=0.2, 
        random_state=RANDOM_STATE, 
        stratify=df['encoded_label']
    )
    
    print(f"Training samples: {len(X_train)}")
    print(f"Validation samples: {len(X_val)}")
    
    # Get class distributions for weighting
    class_counts = df['encoded_label'].value_counts().sort_index()
    total_samples = len(df)
    
    # Calculate class weights inversely proportional to class frequency
    class_weights = torch.tensor(
        [total_samples / (len(class_counts) * count) for count in class_counts],
        dtype=torch.float32,
        device=device
    )
    print(f"Class weights: {class_weights}")
    
    # Load tokenizer
    print("Loading XLM-RoBERTa tokenizer...")
    tokenizer = XLMRobertaTokenizer.from_pretrained('xlm-roberta-base')
    
    # Create datasets and dataloaders with augmentation
    train_dataset = MentalHealthDataset(
        X_train.values, 
        y_train.values, 
        tokenizer, 
        max_length=max_length, 
        augment=use_augmentation
    )
    val_dataset = MentalHealthDataset(
        X_val.values, 
        y_val.values, 
        tokenizer, 
        max_length=max_length
    )
    
    train_dataloader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_dataloader = DataLoader(val_dataset, batch_size=batch_size)
    
    # Initialize model with improvements
    print(f"Initializing improved model (unfreeze_layers={unfreeze_layers}, adapter_size={adapter_size}, dropout={dropout})...")
    model = ImprovedXLMRobertaAdapter(
        num_classes=len(label_encoder.classes_),
        freeze_base=True,
        unfreeze_layers=unfreeze_layers,
        adapter_size=adapter_size,
        dropout=dropout
    )
    model = model.to(device)
    
    # Print number of trainable parameters
    trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
    total_params = sum(p.numel() for p in model.parameters())
    print(f"Trainable parameters: {trainable_params:,} ({trainable_params/total_params:.2%} of total)")
    
    # Initialize optimizer with weight decay for regularization
    no_decay = ['bias', 'LayerNorm.weight']
    optimizer_grouped_parameters = [
        {
            'params': [p for n, p in model.named_parameters() if not any(nd in n for nd in no_decay) and p.requires_grad],
            'weight_decay': 0.01
        },
        {
            'params': [p for n, p in model.named_parameters() if any(nd in n for nd in no_decay) and p.requires_grad],
            'weight_decay': 0.0
        }
    ]
    
    optimizer = AdamW(optimizer_grouped_parameters, lr=learning_rate, eps=1e-8)
    
    # Create learning rate scheduler
    total_steps = len(train_dataloader) * epochs
    scheduler = get_linear_schedule_with_warmup(
        optimizer,
        num_warmup_steps=int(total_steps * 0.1),
        num_training_steps=total_steps
    )
    
    # Initialize loss function
    if use_focal_loss:
        loss_fn = FocalLoss(alpha=class_weights, gamma=2.0)
        print("Using Focal Loss with class weights")
    else:
        loss_fn = nn.CrossEntropyLoss(weight=class_weights)
        print("Using Cross Entropy Loss with class weights")
    
    # Initialize early stopping with longer patience
    early_stopping = EarlyStopping(patience=5, min_delta=0.005)
    
    # Initialize metrics tracking
    train_losses = []
    val_losses = []
    train_accs = []
    val_accs = []
    
    # Training loop
    print(f"Starting training for {epochs} epochs...")
    for epoch in range(epochs):
        print(f"\nEpoch {epoch + 1}/{epochs}")
        
        # Train
        train_loss, train_acc = train_epoch(
            model, train_dataloader, optimizer, scheduler, device,
            loss_fn=loss_fn, accumulation_steps=2
        )
        print(f"Train loss: {train_loss:.4f}, accuracy: {train_acc:.4f}")
        
        # Validate
        val_loss, val_acc, val_preds, val_labels = eval_model(
            model, val_dataloader, device, loss_fn=loss_fn
        )
        print(f"Validation loss: {val_loss:.4f}, accuracy: {val_acc:.4f}")
        
        # Track metrics
        train_losses.append(train_loss)
        val_losses.append(val_loss)
        train_accs.append(train_acc)
        val_accs.append(val_acc)
        
        # Calculate class-wise metrics for current epoch
        if epoch % 2 == 0 or epoch == epochs - 1:
            report = classification_report(
                val_labels, val_preds, 
                target_names=label_encoder.classes_, 
                digits=4,
                output_dict=True
            )
            # Print class-wise F1 scores
            print("Class-wise F1 scores:")
            for cls_name in label_encoder.classes_:
                print(f"  {cls_name}: {report[cls_name]['f1-score']:.4f}")
        
        # Check early stopping
        if early_stopping(val_acc, model):
            print(f"Early stopping triggered after epoch {epoch + 1}")
            break
    
    # Calculate training time
    training_time = time.time() - start_time
    print(f"\nTraining completed in {training_time:.2f} seconds")
    
    # Plot training metrics
    print("\nPlotting training metrics...")
    plot_training_metrics(train_losses, val_losses, train_accs, val_accs, 
                        title='Improved XLM-RoBERTa Training')
    
    # Evaluate final model on validation set
    val_loss, val_acc, val_preds, val_labels = eval_model(model, val_dataloader, device, loss_fn)
    
    # Generate classification report
    print("\nFinal Classification Report:")
    report = classification_report(val_labels, val_preds, 
                                  target_names=label_encoder.classes_, digits=4)
    print(report)
    
    # Plot confusion matrix
    cm = confusion_matrix(val_labels, val_preds)
    plt.figure(figsize=(10, 8))
    sns.heatmap(
        cm, 
        annot=True, 
        fmt='d', 
        cmap='Blues',
        xticklabels=label_encoder.classes_,
        yticklabels=label_encoder.classes_
    )
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Confusion Matrix')
    plt.savefig(f"{output_dir}/confusion_matrix.png")
    plt.show()
    
    # Save model
    print(f"\nSaving model to {output_dir}...")
    torch.save(model.state_dict(), f"{output_dir}/model.pt")
    
    # Save tokenizer
    tokenizer.save_pretrained(output_dir)
    
    # Save label encoder
    with open(f"{output_dir}/label_encoder.pkl", 'wb') as f:
        pickle.dump(label_encoder, f)
    
    # Save training config
    config = {
        'max_length': max_length,
        'unfreeze_layers': unfreeze_layers,
        'adapter_size': adapter_size,
        'dropout': dropout,
        'num_classes': len(label_encoder.classes_),
        'class_names': label_encoder.classes_.tolist(),
        'training_time': training_time,
        'final_val_accuracy': val_acc,
        'epochs_trained': len(train_losses),
        'class_weights': class_weights.cpu().numpy().tolist()
    }
    
    with open(f"{output_dir}/config.pkl", 'wb') as f:
        pickle.dump(config, f)
    
    print("Model and configuration saved successfully")
    
    # Return results
    return {
        'model': model,
        'tokenizer': tokenizer,
        'label_encoder': label_encoder,
        'metrics': {
            'train_losses': train_losses,
            'val_losses': val_losses,
            'train_accs': train_accs,
            'val_accs': val_accs,
            'final_val_acc': val_acc,
            'classification_report': report
        },
        'config': config
    }

# Function to load model and make predictions
def load_model_and_predict(text, model_dir):
    """
    Load saved model and make a prediction
    """
    # Load config
    with open(f"{model_dir}/config.pkl", 'rb') as f:
        config = pickle.load(f)
    
    # Load label encoder
    with open(f"{model_dir}/label_encoder.pkl", 'rb') as f:
        label_encoder = pickle.load(f)
    
    # Load tokenizer
    tokenizer = XLMRobertaTokenizer.from_pretrained(model_dir)
    
    # Initialize model
    model = ImprovedXLMRobertaAdapter(
        num_classes=config['num_classes'],
        unfreeze_layers=config.get('unfreeze_layers', 2),
        adapter_size=config.get('adapter_size', 128),
        dropout=config.get('dropout', 0.4)
    )
    
    # Load model weights
    model.load_state_dict(torch.load(f"{model_dir}/model.pt", map_location=device))
    model = model.to(device)
    
    # Make prediction
    result = predict_mental_health(
        text, tokenizer, model, label_encoder, device,
        max_length=config['max_length']
    )
    
    return result

# Example usage for Jupyter notebook

# Set your parameters here
data_path = "../Data/Preprocessed_Manual_Labelled_Dataset .csv"
output_dir = "./improved_xlm_roberta_model"

# Run the training with improved parameters
results = train_improved_transformer(
    data_path=data_path,
    output_dir=output_dir,
    unfreeze_layers=2,            # Unfreeze more layers for better adaptation
    adapter_size=128,             # Larger adapter for more capacity
    dropout=0.4,                  # Higher dropout to prevent overfitting
    epochs=15,                    # Train longer with early stopping
    batch_size=16,                # Adjust based on memory
    learning_rate=1e-5,           # Lower learning rate for stability
    use_focal_loss=True,          # Use focal loss for imbalanced classes
    use_augmentation=True         # Use text augmentation
)
```

```python
from sklearn.metrics import (accuracy_score, precision_recall_fscore_support, 
                            roc_auc_score, classification_report, confusion_matrix,
                            roc_curve, precision_recall_curve, auc)
from sklearn.calibration import calibration_curve
import seaborn as sns
import matplotlib.pyplot as plt
from lime.lime_text import LimeTextExplainer
from IPython.display import HTML, display
# Import the required libraries
from sklearn.metrics import roc_curve, precision_recall_curve, auc
# Try to import calibration_curve from the correct location
try:
    from sklearn.metrics import calibration_curve
except ImportError:
    try:
        # In older versions, it was in sklearn.calibration
        from sklearn.calibration import calibration_curve
    except ImportError:
        # Define a simple version if we can't import it
        def calibration_curve(y_true, y_prob, n_bins=5):
            """Simple implementation of calibration curve"""
            bins = np.linspace(0., 1. + 1e-8, n_bins + 1)
            binids = np.digitize(y_prob, bins) - 1
            bin_sums = np.bincount(binids, weights=y_prob, minlength=len(bins))
            bin_true = np.bincount(binids, weights=y_true, minlength=len(bins))
            bin_total = np.bincount(binids, minlength=len(bins))
            nonzero = bin_total != 0
            prob_true = bin_true[nonzero] / bin_total[nonzero]
            prob_pred = bin_sums[nonzero] / bin_total[nonzero]
            return prob_true, prob_pred
```

```python
# 1. Extract model and data from results
model = results['model']
tokenizer = results['tokenizer']
label_encoder = results['label_encoder']

# 2. Load the dataset and create the test set
df = pd.read_csv(data_path)
df['preprocessed_text'] = df['cleaned_text'].apply(preprocess_sinhala_text)

# Apply the label encoder
df['y'] = label_encoder.transform(df['label'])

# Split the data
_, X_test, _, y_test = train_test_split(
    df['preprocessed_text'], 
    df['y'],
    test_size=0.2, 
    random_state=RANDOM_STATE, 
    stratify=df['y']
)

# Create a test dataset and dataloader
test_dataset = MentalHealthDataset(
    X_test.values, 
    y_test.values, 
    tokenizer, 
    max_length=128
)
test_dataloader = DataLoader(test_dataset, batch_size=16)

# Use the same loss function
loss_fn = FocalLoss(
    alpha=torch.tensor(results['config']['class_weights'], dtype=torch.float32).to(device),
    gamma=2.0
)

# 3. Run evaluation to get predictions
test_loss, test_acc, y_test_pred, y_test_true = eval_model(
    model, test_dataloader, device, loss_fn
)

# 4. Get prediction probabilities
y_proba = []
model.eval()
with torch.no_grad():
    for batch in tqdm(test_dataloader, desc="Getting probabilities"):
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        
        # Forward pass
        logits = model(input_ids=input_ids, attention_mask=attention_mask)
        probs = torch.softmax(logits, dim=1).cpu().numpy()
        y_proba.extend(probs)

y_proba = np.array(y_proba)


# 7. LIME explanation for interpretability
def predict_fn_for_lime(texts):
    """Wrapper function for model prediction that LIME can use"""
    batch_probs = []
    
    for text in texts:
        # Preprocess
        text = preprocess_sinhala_text(text)
        
        # Tokenize
        encoding = tokenizer(
            text,
            add_special_tokens=True,
            max_length=128,
            padding='max_length',
            truncation=True,
            return_tensors='pt'
        )
        
        # Move to device
        input_ids = encoding['input_ids'].to(device)
        attention_mask = encoding['attention_mask'].to(device)
        
        # Get prediction
        model.eval()
        with torch.no_grad():
            logits = model(input_ids=input_ids, attention_mask=attention_mask)
            probs = torch.softmax(logits, dim=1).cpu().numpy()[0]
        
        batch_probs.append(probs)
    
    return np.array(batch_probs)

# Choose an example from test set
idx = 5  # You can change this to any index within the range of X_test
example_text = X_test.iloc[idx]
true_class = label_encoder.inverse_transform([y_test.iloc[idx]])[0]

print(f"Analyzing example text (index {idx}):")
print(f"True class: {true_class}")
print(f"Text: {example_text[:100]}...")  # Show first 100 chars

# Create explainer
explainer = LimeTextExplainer(class_names=label_encoder.classes_, split_expression=r"\s+")

# Generate explanation
exp = explainer.explain_instance(
    example_text, 
    predict_fn_for_lime,
    num_features=10, 
    num_samples=300
)

# Show prediction
predicted_probs = predict_fn_for_lime([example_text])[0]
predicted_class = label_encoder.classes_[np.argmax(predicted_probs)]
print(f"Predicted class: {predicted_class} (confidence: {max(predicted_probs):.3f})")

# Display explanation
display(HTML(exp.as_html()))

# Save explanation visualization
exp.save_to_file('lime_explanation_transformer.html')

# 7. ROC curves and AUC calculation
# One-hot encode test labels for ROC curves
y_bin = np.zeros((len(y_test_true), len(label_encoder.classes_)))
for i, y_true in enumerate(y_test_true):
    y_bin[i, y_true] = 1

colors = ["navy", "darkorange", "green"][:len(label_encoder.classes_)]

# Plot ROC curves
plt.figure(figsize=(10, 8))
for i, color in enumerate(colors):
    fpr, tpr, _ = roc_curve(y_bin[:, i], y_proba[:, i])
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, color=color, lw=2,
             label=f'{label_encoder.classes_[i]} (AUC = {roc_auc:.2f})')

plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves')
plt.legend(loc="lower right")
plt.grid(True)
plt.savefig("roc_curves_transformer.png")
plt.show()

# 8. Precision-Recall curves
plt.figure(figsize=(10, 8))
for i, color in enumerate(colors):
    p, r, _ = precision_recall_curve(y_bin[:, i], y_proba[:, i])
    pr_auc = auc(r, p)
    plt.plot(r, p, color=color, lw=2,
             label=f'{label_encoder.classes_[i]} (AUC = {pr_auc:.2f})')

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curves')
plt.legend(loc="lower left")
plt.grid(True)
plt.savefig("pr_curves_transformer.png")
plt.show()

# 9. Calibration curves
plt.figure(figsize=(10, 8))
for i, color in enumerate(colors):
    fraction_of_positives, mean_predicted_value = calibration_curve(
        y_bin[:, i], y_proba[:, i], n_bins=10)
    plt.plot(mean_predicted_value, fraction_of_positives, "s-",
             color=color, label=label_encoder.classes_[i])

plt.plot([0, 1], [0, 1], 'k--', label='Perfectly calibrated')
plt.xlabel('Mean predicted probability')
plt.ylabel('Fraction of positives')
plt.title('Calibration Curves')
plt.legend(loc="upper left")
plt.grid(True)
plt.savefig("calibration_curves_transformer.png")
plt.show()



print("\nAll evaluation visualizations have been saved as PNG files.")
```
